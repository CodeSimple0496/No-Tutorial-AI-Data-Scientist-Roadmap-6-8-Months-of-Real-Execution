```python
## Operators 

"""operators are the symbols we use to perform actions on variables and data. Here are the most common ones categorized by what they do."""


# 1. Arithmatic Operators

"""
Arithmetic operators are used to perform mathematical operations:

1. Addition (+) - Adds two numbers
2. Subtraction (-) - Subtracts one number from another
3. Multiplication (*) - Multiplies two numbers
4. Division (/) - Divides one number by another (returns float)
5. Floor Division (//) - Divides and returns the integer part
6. Modulus (%) - Returns the remainder after division
7. Exponentiation (**) - Raises a number to the power of another
"""

# Examples of Arithmetic Operators

a = 10
b = 3

print(f"a = {a}, b = {b}")
print(f"Addition: a + b = {a + b}")           # 13
print(f"Subtraction: a - b = {a - b}")       # 7
print(f"Multiplication: a * b = {a * b}")    # 30
print(f"Division: a / b = {a / b}")          # 3.333...
print(f"Floor Division: a // b = {a // b}")  # 3
print(f"Modulus: a % b = {a % b}")           # 1
print(f"Exponentiation: a ** b = {a ** b}")  # 1000
```

```python
# 2. Comparison Operators

"""
Comparison operators are used to compare two values and return True or False (Boolean):

1. Equal to (==) - Checks if two values are equal
2. Not equal to (!=) - Checks if two values are not equal
3. Greater than (>) - Checks if left value is greater than right value
4. Less than (<) - Checks if left value is less than right value
5. Greater than or equal to (>=) - Checks if left value is greater than or equal to right value
6. Less than or equal to (<=) - Checks if left value is less than or equal to right value
"""

# Examples of Comparison Operators

x = 15
y = 8

print(f"x = {x}, y = {y}")
print(f"Equal to: x == y is {x == y}")              # False
print(f"Not equal to: x != y is {x != y}")          # True
print(f"Greater than: x > y is {x > y}")            # True
print(f"Less than: x < y is {x < y}")               # False
print(f"Greater or equal: x >= y is {x >= y}")      # True
print(f"Less or equal: x <= y is {x <= y}")         # False
```

```python
# 3. Logical Operators

"""
Logical operators are used to combine conditional statements and return True or False:

1. and - Returns True if both conditions are True
2. or - Returns True if at least one condition is True
3. not - Reverses the boolean result (True becomes False, False becomes True)
"""

# Examples of Logical Operators

p = 20
q = 10

print(f"p = {p}, q = {q}")
print()

# AND operator
print(f"p > 15 and q > 5: {p > 15 and q > 5}")           # True (both conditions are True)
print(f"p > 15 and q > 15: {p > 15 and q > 15}")         # False (second condition is False)
print()

# OR operator
print(f"p > 25 or q > 5: {p > 25 or q > 5}")             # True (second condition is True)
print(f"p > 25 or q > 25: {p > 25 or q > 25}")           # False (both conditions are False)
print()

# NOT operator
print(f"not (p > 15): {not (p > 15)}")                   # False (p > 15 is True, not reverses it)
print(f"not (p > 25): {not (p > 25)}")                   # True (p > 25 is False, not reverses it)
```

```python
# 4. Assignment Operators

"""
Assignment operators are used to assign values to variables and perform operations in one step:

1. = - Assigns a value to a variable
2. += - Adds and assigns (x += 5 is same as x = x + 5)
3. -= - Subtracts and assigns (x -= 5 is same as x = x - 5)
4. *= - Multiplies and assigns (x *= 5 is same as x = x * 5)
5. /= - Divides and assigns (x /= 5 is same as x = x / 5)
6. //= - Floor divides and assigns (x //= 5 is same as x = x // 5)
7. %= - Modulus and assigns (x %= 5 is same as x = x % 5)
8. **= - Exponentiation and assigns (x **= 5 is same as x = x ** 5)
"""

# Examples of Assignment Operators

num = 10
print(f"Initial value: num = {num}")
print()

# +=
num += 5
print(f"After num += 5: num = {num}")  # 15

# -=
num -= 3
print(f"After num -= 3: num = {num}")  # 12

# *=
num *= 2
print(f"After num *= 2: num = {num}")  # 24

# /=
num /= 4
print(f"After num /= 4: num = {num}")  # 6.0

# //=
num //= 2
print(f"After num //= 2: num = {num}")  # 3.0

# %=
num %= 2
print(f"After num %= 2: num = {num}")  # 1.0

# **=
num **= 3
print(f"After num **= 3: num = {num}")  # 1.0
```

```python
# 5. Identity Operators

"""
Identity operators are used to compare objects, not if they are equal, but if they are the same object with the same memory location:

1. is - Returns True if both variables are the same object
2. is not - Returns True if variables are not the same object
"""

# Examples of Identity Operators

list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print("Identity Operators:")
print(f"list1 = {list1}")
print(f"list2 = {list2}")
print(f"list3 = list1")
print()
print(f"list1 is list2: {list1 is list2}")      # False (different objects, even though contents are same)
print(f"list1 is list3: {list1 is list3}")      # True (same object reference)
print(f"list1 is not list2: {list1 is not list2}")  # True
print()
print()

# 6. Membership Operators

"""
Membership operators are used to test if a value exists in a sequence (like lists, strings, tuples, etc.):

1. in - Returns True if a value exists in a sequence
2. not in - Returns True if a value does not exist in a sequence
"""

# Examples of Membership Operators

fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
text = "Hello World"

print("Membership Operators:")
print(f"fruits = {fruits}")
print(f"numbers = {numbers}")
print(f"text = '{text}'")
print()

# in operator
print(f"'apple' in fruits: {'apple' in fruits}")           # True
print(f"'orange' in fruits: {'orange' in fruits}")         # False
print(f"3 in numbers: {3 in numbers}")                     # True
print(f"10 in numbers: {10 in numbers}")                   # False
print(f"'Hello' in text: {'Hello' in text}")               # True
print()

# not in operator
print(f"'grape' not in fruits: {'grape' not in fruits}")   # True
print(f"'apple' not in fruits: {'apple' not in fruits}")   # False
print(f"100 not in numbers: {100 not in numbers}")         # True
print(f"'xyz' not in text: {'xyz' not in text}")           # True
```

```python
# 7. Bitwise Operators

"""
Bitwise operators are used to perform operations on binary representations of integers.
They work on the bits of the numbers:

1. & (AND) - Sets each bit to 1 if both bits are 1
2. | (OR) - Sets each bit to 1 if one or both bits are 1
3. ^ (XOR) - Sets each bit to 1 if only one of the bits is 1
4. ~ (NOT) - Inverts all the bits
5. << (Left Shift) - Shifts bits to the left, filling with 0s on the right
6. >> (Right Shift) - Shifts bits to the right, discarding rightmost bits
"""

# Examples of Bitwise Operators

a = 12  # Binary: 1100
b = 10  # Binary: 1010

print("Bitwise Operators:")
print(f"a = {a} (Binary: {bin(a)})")
print(f"b = {b} (Binary: {bin(b)})")
print()

# & (AND)
print(f"a & b = {a & b} (Binary: {bin(a & b)})")  # 8 (Binary: 1000)
print()

# | (OR)
print(f"a | b = {a | b} (Binary: {bin(a | b)})")  # 14 (Binary: 1110)
print()

# ^ (XOR)
print(f"a ^ b = {a ^ b} (Binary: {bin(a ^ b)})")  # 6 (Binary: 0110)
print()

# ~ (NOT) - Note: inverts all bits including sign
print(f"~a = {~a} (Binary: {bin(~a & 0xFF)})")    # -13
print()

# << (Left Shift) - Shifts left by 2 positions
print(f"a << 2 = {a << 2} (Binary: {bin(a << 2)})")  # 48 (Binary: 110000)
print()

# >> (Right Shift) - Shifts right by 2 positions
print(f"a >> 2 = {a >> 2} (Binary: {bin(a >> 2)})")  # 3 (Binary: 0011)
```

## Conditional Statement


```python
# Conditional Statements

"""
Conditional statements allow you to execute different blocks of code based on different conditions.
They control the flow of your program.

1. if - Executes code if a condition is True
2. elif - Executes code if the previous condition is False and this condition is True
3. else - Executes code if all previous conditions are False
"""

# Example 1: Simple if statement
age = 20
print("Example 1: Simple if statement")
if age >= 18:
    print(f"You are {age} years old. You are an adult!")
print()

# Example 2: if-else statement
score = 45
print("Example 2: if-else statement")
if score >= 50:
    print(f"Your score is {score}. You PASSED!")
else:
    print(f"Your score is {score}. You FAILED!")
print()

# Example 3: if-elif-else statement
marks = 85
print("Example 3: if-elif-else statement")
if marks >= 90:
    grade = "A"
elif marks >= 80:
    grade = "B"
elif marks >= 70:
    grade = "C"
elif marks >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Your marks: {marks}, Grade: {grade}")
print()

# Example 4: Nested if statements
temperature = 25
print("Example 4: Nested if statements")
if temperature > 0:
    if temperature > 20:
        print(f"Temperature is {temperature}°C. It's warm!")
    else:
        print(f"Temperature is {temperature}°C. It's cool!")
else:
    print(f"Temperature is {temperature}°C. It's freezing!")
print()

# Example 5: Multiple conditions with logical operators
number = 15
print("Example 5: Multiple conditions with logical operators")
if number > 0 and number < 20:
    print(f"{number} is a positive number between 0 and 20")
elif number > 20 or number < -20:
    print(f"{number} is outside the range of -20 to 20")
else:
    print(f"{number} is between -20 and 0")
print()

# Example 6: Ternary operator (one-line if-else)
x = 10
print("Example 6: Ternary operator")
result = "Even" if x % 2 == 0 else "Odd"
print(f"{x} is {result}")
```
