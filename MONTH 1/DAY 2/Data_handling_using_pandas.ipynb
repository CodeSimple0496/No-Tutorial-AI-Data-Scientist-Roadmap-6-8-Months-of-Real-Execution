## 1. Learn Basic Pandas Operations (Their given below keyword uses to Overview the Data's Health.)

1️⃣ Dataset Overview

How many rows and columns are in the dataset?

What are the data types of each column?

Which columns are categorical and which are numerical?

```
# Import libraries
import pandas as pd

# Load Dataset
df = pd.read_csv('/content/Employee.csv')

# Find out the data Overview

print("Access the top 5 rows")
display(df.head())  # Display the first few rows(5 rows) of the DataFrame
print("Accss the last 5 rows")
display(df.tail())  # Display the last few rows(5 rows) of the DataFrame

df.info()  # Get a summary of the DataFrame
display(df.describe())  # Get statistical summary of numerical columns


# Access 273 to 489 row in the data.
display(df[273:489])

# total size of the dataset
df.shape
```

- Handling Missing Data

Drop missing rows

Fill missing values

Drop the Duplicate rows

Compare before vs after

```
# 1. Drop missing rows
df = df.dropna()

# 2. FIll the missing values
df = df.fillna(0)

# 3. Drop the Duplicate rows
df = df.drop_duplicates().reset_index(drop=True)

df.shape
# After the Handling data we find out many duplicate data so it have provide more accuracy.
```

```
# Get the number of rows and columns
print(f"1. The dataset has {df.shape[0]} rows and {df.shape[1]} columns.")

# Get and print the data types of each column
print("\n2. Data types of each column:")
print(df.dtypes)

# Identify categorical columns (usually 'object' type in pandas)
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Identify numerical columns (usually 'int64' or 'float64' types)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Print the identified categorical and numerical columns
print(f"\nCategorical columns: {categorical_cols}")
print(f"Numerical columns: {numerical_cols}")
```

2️⃣ Basic Statistics

What is the average age of employees?

What is the minimum and maximum age?

What is the average experience in the current domain?

```
# Average Age of Employees.
average_age = df['Age'].mean()
print(f"1. The average age of employees is {average_age:.2f} years.")

# Min and Max Age
min_age = df['Age'].min()
max_age = df['Age'].max()
print(f"2. The minimum age of employees is {min_age} years and the maximum age is {max_age} years.")

# Average Experience
average_exp = df['ExperienceInCurrentDomain'].mean()
print(f"3. The average experience in the current domain is {average_exp:.2f} years.")
```

3️⃣ Filtering Data

Show employees older than 35.

Show employees with PaymentTier = 3.

Show female employees who left the company.

```
print("Employees older than 35:")
older_then_35 = df['Age'] > 35
display(df[older_then_35])

print("Employees with PaymentTier = 3:")
payment_tire_3 = df['PaymentTier'] == 3
display(df[payment_tire_3])

print("Female employees who left the company:")
left_female_emp = (df['Gender'] == 'Female') & (df['LeaveOrNot'] == 0)
display(df[left_female_emp])
```

4️⃣ Sorting Data

Sort employees by Age (descending).

Sort employees by ExperienceInCurrentDomain and Age together.

```
# Sort Employees Age By descending order.
print("Sort Employees Age By descending order.")
sorted_age = df.sort_values(by='Age', ascending=False)
display(sorted_age)


# Sort employees Experience and Age Together.
print('Sort employees by ExperienceInCurrentDomain and Age together')
sort_Exp_and_Age = df.sort_values(by = ['ExperienceInCurrentDomain', 'Age'], ascending=False)
display(sort_Exp_and_Age)
```

⚙️ Level 2 — Business-Oriented Pandas Questions

5️⃣ GroupBy Analysis

Average age by gender.

Count employees in each city.

Average PaymentTier by education level.

Attrition count by city (LeaveOrNot).

```
# Average age by gender
Avg_age_by_gender = df.groupby('Gender')['Age'].mean()
print("Average age by gender:")
display(Avg_age_by_gender)

# Count employees in each city
emp_count_by_city = df['City'].value_counts()
print("\nCount employees in each city:")
display(emp_count_by_city)

# Average paymentTire by education level
avg_PT_by_edu = df.groupby('Education')['PaymentTier'].mean()
print("\nAverage paymentTire by education level:")
display(avg_PT_by_edu)

# Attrition Count by City (LeaveOrNot)
attrition_count_by_city = df.groupby(['City', 'LeaveOrNot']).size()
print("\nAttrition count by city (LeaveOrNot):")
display(attrition_count_by_city)
```

6️⃣ Attrition Analysis

What percentage of employees left the company?

Which gender has higher attrition?

Which education level has highest attrition?

Does PaymentTier affect leaving?

```
# Percentage of the employee left the company
per_emp_left = (df['LeaveOrNot'] == 1).mean() * 100
print(f"1. The percentage of employees who left the company is {per_emp_left}")

# Which gender has higher attrition
gender_attrition = df.groupby('Gender')['LeaveOrNot'].mean()
print("\n2. Attrition rate by gender:")
display(gender_attrition)

# Which Education level has highest attrition
edu_attrition = df.groupby('Education')['LeaveOrNot'].mean()
print("\n3. Attrition rate by Education:")
display(edu_attrition)

#Does PaymentTire affect leaving?
payment_attrition = df.groupby('PaymentTier')['LeaveOrNot'].mean()
print("\n4. Attrition rate by PaymentTier:")
display(payment_attrition)

## Yes, Payment is affect the Leaving the employees in Company. for the advancing the upper position.(Specially Tire-2)
```

7️⃣ Experience Insights

Average experience of employees who left vs stayed.

Does low experience correlate with leaving?

```
# Average experience of employees who left vs stayed.
avg_exp_by_leave = df.groupby('LeaveOrNot')['ExperienceInCurrentDomain'].mean()
print("Average experience of employees who left vs stayed:")
display(avg_exp_by_leave)


# Does low experience correlate with leaving?
exp_corr_leav = df.groupby('ExperienceInCurrentDomain')['LeaveOrNot'].mean()
print("Does low experience correlate with leaving:")
display(exp_corr_leav)
```
