```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('/content/Employee.csv')


# Remove the duplicates
df.drop_duplicates(inplace=True)
display(df.head())

# Set the style for better visualization
sns.set(style="whitegrid")  # it is used to better reading and undersatnd of the Visualization
```

### 1. Histogram: Distribution of Age

A histogram displays the distribution of a numerical variable. It divides the data into 'bins' and counts how many data points fall into each bin, representing these counts with bars.

Here's what each part of the code does:

*   **`plt.figure(figsize=(8, 6))`**: Sets the size of the plot to 8 inches wide and 6 inches tall.
*   **`sns.histplot(data=df, x='Age', bins=15, kde=True, color='skyblue')`**: This is the core command to create the histogram.
    *   `data=df`: Specifies that the data comes from your DataFrame `df`.
    *   `x='Age'`: Uses the 'Age' column for the horizontal x-axis.
    *   `bins=15`: Divides the 'Age' data into 15 distinct groups or intervals.
    *   `kde=True`: Adds a Kernel Density Estimate (KDE) line, which is a smoothed representation of the distribution.
    *   `color='skyblue'`: Sets the color of the histogram bars.
*   **`plt.title('Distribution of Age')`**: Adds a main title to the plot.
*   **`plt.xlabel('Age')`** and **`plt.ylabel('Frequency')`**: Label the x and y axes for clarity.
*   **`plt.show()`**: Displays the generated plot.

```
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=15, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

### 2. Bar Chart: Average Payment Tier by Education

A bar chart is used to display and compare the quantities of different categories. In this case, we're comparing the average payment tier across different education levels.

Here's a breakdown of the code:

*   **`avg_payment = df.groupby('Education')['PaymentTier'].mean().reset_index().sort_values(by='PaymentTier')`**: This line first calculates the average `PaymentTier` for each `Education` level, then converts the result into a DataFrame, and finally sorts it by the `PaymentTier`.
*   **`plt.figure(figsize=(8, 6))`**: Sets the plot size.
*   **`sns.barplot(data=avg_payment, x='Education', y='PaymentTier', palette='viridis', order=avg_payment['Education'])`**: This creates the bar chart.
    *   `data=avg_payment`: Uses the `avg_payment` DataFrame.
    *   `x='Education'`: Places education levels on the x-axis.
    *   `y='PaymentTier'`: Represents the average payment tier on the y-axis.
    *   `palette='viridis'`: Applies a color palette to the bars.
    *   `order=avg_payment['Education']`: Ensures the bars are ordered according to the sorted `avg_payment` DataFrame.
*   **`plt.title('Average Payment Tier by Education')`**: Adds a title.
*   **`plt.xlabel('Education')`** and **`plt.ylabel('Average Payment Tier')`**: Labels the axes.
*   **`plt.show()`**: Displays the plot.

```
avg_payment = df.groupby('Education')['PaymentTier'].mean().reset_index().sort_values(by='PaymentTier')
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_payment, x='Education', y='PaymentTier', hue='Education', palette='viridis', order=avg_payment['Education'], legend=False)
plt.title('Average Payment Tier by Education')
plt.xlabel('Education')
plt.ylabel('Average Payment Tier')
plt.show()

```

### 4. Count Plot: Distribution of Cities

A count plot shows the counts of observations in each category using bars. It's similar to a histogram but for categorical variables.

Here's a breakdown of the code:

*   **`plt.figure(figsize=(10, 6))`**: Sets the size of the plot.
*   **`sns.countplot(data=df, y='City', order=df['City'].value_counts().index, hue='City', palette='coolwarm', legend=False)`**: This function shows the counts of observations in each category.
    *   `data=df`: Specifies the DataFrame.
    *   `y='City'`: Sets the 'City' column on the y-axis, resulting in horizontal bars.
    *   `order=df['City'].value_counts().index`: Sorts the cities by their frequency in descending order.
    *   `hue='City'`: Assigns a different color to each city category, resolving the `FutureWarning`.
    *   `palette='coolwarm'`: Applies a color palette.
    *   `legend=False`: Hides the legend as the `hue` variable is also on the y-axis.
*   **`plt.title('Distribution of Employees by City')`**: Sets the plot title.
*   **`plt.xlabel('Number of Employees')`** and **`plt.ylabel('City')`**: Labels the axes.
*   **`plt.show()`**: Displays the plot.

```
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='City', order=df['City'].value_counts().index, hue='City', palette='coolwarm', legend=False)
plt.title('Distribution of Employees by City')
plt.xlabel('Number of Employees')
plt.ylabel('City')
plt.show()
```

## 4. Correlation Analysis

To understand the relationships between different features in the dataset, especially how they correlate with 'LeaveOrNot', we perform a correlation analysis. Since correlation matrices typically work with numerical data, we first need to convert our categorical features into numerical representations using **Label Encoding**.

### 1. Convert Categorical Columns to Numbers (Label Encoding)

Label Encoding assigns a unique integer to each category in a categorical column. This allows these columns to be included in numerical calculations like correlation.

Here's a breakdown of the code:

*   **`from sklearn.preprocessing import LabelEncoder`**: Imports the `LabelEncoder` class from scikit-learn.
*   **`le = LabelEncoder()`**: Initializes a `LabelEncoder` object.
*   **`df_encoded = df.copy()`**: Creates a copy of the original DataFrame to avoid modifying it directly.
*   **`for col in df.select_dtypes(include='object').columns:`**: Iterates through each column in the copied DataFrame that has an 'object' (string/categorical) data type.
*   **`df_encoded[col] = le.fit_transform(df_encoded[col])`**: Applies label encoding to the current categorical column. `fit_transform` learns the unique categories and then transforms them into numerical labels.

```
from sklearn.preprocessing import LabelEncoder

# Create a copy of the DataFrame to encode categorical variables
df_encoded = df.copy()

# Apply Label Encoding to categorical columns
for col in df_encoded.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

display(df_encoded.head())
```

### 2. Create Correlation Matrix

A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables. The value ranges from -1 to 1, where:
*   1 indicates a perfectly positive correlation.
*   -1 indicates a perfectly negative correlation.
*   0 indicates no linear correlation.

Here's what the code does:

*   **`correlation_matrix = df_encoded.corr()`**: Calculates the pairwise correlation of all columns in the `df_encoded` DataFrame. By default, this uses Pearson correlation.

```
# Calculate the correlation matrix
correlation_matrix = df_encoded.corr()
display(correlation_matrix.head())
```

### 3. Plot Heatmap

A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. It's an excellent way to visualize a correlation matrix, making it easy to spot strong positive or negative correlations.

Here's a breakdown of the code:

*   **`plt.figure(figsize=(12, 10))`**: Sets the size of the plot to ensure readability.
*   **`sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)`**: Creates the heatmap.
    *   `correlation_matrix`: The data to be plotted.
    *   `annot=True`: Displays the correlation values on the heatmap cells.
    *   `cmap='coolwarm'`: Sets the color map. 'coolwarm' is good for showing positive (warm colors) and negative (cool colors) correlations.
    *   `fmt='.2f'`: Formats the annotation text to two decimal places.
    *   `linewidths=.5`: Adds lines between cells for better visual separation.
*   **`plt.title('Correlation Matrix of Employee Dataset')`**: Adds a title to the heatmap.
*   **`plt.show()`**: Displays the plot.

```
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Employee Dataset')
plt.show()
```

### What is data visualization?

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data. It helps in making complex data understandable and usable for decision-making.

### Why graphs are better than tables?

Graphs are often better than tables for several reasons:
1.  **Easier Comprehension**: Visual patterns, trends, and relationships are much easier and quicker to spot in a graph than by scanning rows and columns of numbers in a table.
2.  **Impact and Engagement**: Graphs can tell a story more effectively and engage an audience more than raw data. They make data more memorable.
3.  **Identification of Outliers and Anomalies**: Outliers and anomalies often stand out immediately in a visual representation, which might be missed in a table.
4.  **Comparison and Relationships**: Graphs excel at showing comparisons between different categories or the relationship between two or more variables.
